```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# データの準備
chunk_sizes = [100, 200, 500, 1000]
top_ks = [20, 10, 4]

data = np.array([
    [0.90, 0.85, 0.70],  # chunk_size 100 の結果
    [0.95, 0.85, 0.65],  # chunk_size 200 の結果
    [0.85, 0.80, 0.85],  # chunk_size 500 の結果
    [0.80, 0.75, 0.70]   # chunk_size 1000 の結果
]).T  

# ヒートマップの作成
plt.figure(figsize=(10, 8))
sns.heatmap(data,
            annot=True,
            fmt='.2f',
            cmap='YlOrRd',
            xticklabels=chunk_sizes,  # chunk_sizesをx軸に
            yticklabels=top_ks)       # top_ksをy軸に

plt.xlabel('Chunk Size')  
plt.ylabel('Top K')       
plt.title('Context Recall Heatmap')

plt.tight_layout()
plt.show()
```
